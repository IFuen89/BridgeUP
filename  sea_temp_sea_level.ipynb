```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SeaTemp_SeaLevel_anomaly.csv')
```

```python
Year = df["Year"]
STemp = df["Sea temperature anomaly (C)"]
CIMean = df["Change in mean sea level (mm)"]
```

```python
plt.xlabel("Years")
plt.plot(Year, STemp, color = "Red")
plt.ylabel("Sea Temperature")
plt.twinx()
plt.plot(Year, CIMean, color = "Blue")
plt.ylabel("Sea level Mean")
plt.show()
```
